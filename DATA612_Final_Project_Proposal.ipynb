# Final Project Proposal

## Scalable Movie Recommendation Using Apache Spark: An Item-Based Collaborative Filtering Approach

## Project Overview

For my final project, I plan to build a scalable movie recommender system using the MovieLens 1M dataset and Apache Spark. The system will use an Item-Based Collaborative Filtering approach to generate personalized movie recommendations based on user rating behavior.

Unlike my previous course projects that used small synthetic datasets, this project will use a real-world dataset containing approximately one million movie ratings. This will allow me to evaluate how the recommender system performs on a much larger dataset while exploring the advantages and tradeoffs of distributed computing.

In addition to building the recommender system, I plan to evaluate how well the system performs as the amount of data increases. This analysis will help identify when Apache Spark becomes a better solution than using a single computer to process the data.

The overall goal of this project is not only to build an effective recommender system, but also to better understand when the benefits of using Apache Spark outweigh the extra setup and effort required to use it.

## Dataset

For this project, I plan to use the MovieLens 1M dataset provided by the GroupLens Research Group at the University of Minnesota. The dataset contains over one million movie ratings from about 6,000 users across roughly 4,000 movies.

I selected this dataset because it is a well-known benchmark for recommender system research and provides enough data to explore how Apache Spark performs with a much larger recommendation dataset. The size of the dataset also creates an opportunity to evaluate when distributed computing becomes more beneficial than processing everything on a single computer.

The dataset includes separate files for users, movies, and ratings, providing an opportunity to combine multiple data sources while building the recommender system.

## Recommendation Method

This project will use an Item-Based Collaborative Filtering recommender system to recommend movies based on similar rating patterns from users. Instead of recommending movies because they share the same genre or storyline, the recommendations are based on how groups of users rated those movies.

I selected this approach because it allows me to build on several concepts I learned throughout this course, including collaborative filtering, similarity calculations, recommendation evaluation, and Apache Spark. Rather than learning a completely new recommendation method, I want to apply what I have already learned to a much larger, real-world dataset.

This approach also gives me the opportunity to see how well the recommender performs as the amount of data increases and to better understand when Apache Spark becomes a better choice than using a single computer.

## Evaluation Plan

To evaluate the recommender system, I plan to focus on both the quality of the recommendations and how the system performs as the amount of data increases.

First, I will evaluate whether the recommender produces meaningful movie recommendations by applying evaluation techniques learned throughout the course.

Next, I will compare how the system performs when working with different amounts of data. Rather than focusing only on recommendation results, I also want to observe how processing time changes as the dataset grows.

The goal is to better understand when Apache Spark becomes a better choice than using a single computer and what tradeoffs exist between simplicity and scalability.

One question I hope to answer through this project is when processing the recommendations becomes too demanding for a single computer. I will look for signs such as longer processing times, increased memory usage, or slower recommendation generation as the amount of data increases. This will help me better understand when Apache Spark provides a practical advantage.

## Expected Outcomes

By completing this project, I expect to gain a better understanding of how recommender systems perform when working with large, real-world datasets. I also expect to strengthen my understanding of Apache Spark by applying it to a project that goes beyond the small examples used earlier in the course.

I hope to better understand the tradeoffs between using a traditional single computer and a distributed computing framework such as Spark. Rather than assuming Spark is always the better solution, I want to determine when the benefits of using Spark outweigh the extra setup and effort required to use it.

Finally, I expect this project to strengthen my understanding of collaborative filtering while giving me practical experience designing, evaluating, and improving a recommender system using a large real-world dataset.
